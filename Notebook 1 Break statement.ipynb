```python
#Break statement in python is a loop control statement .it is used to control the sequence of the loop
#1
num = 0
for i in range (10):
    num += 1
    if num == 8:
        break 
print("The num has value:",num)
print("out of loop")
```

```python
# using for loop
s = 'geeksforgeeks'
for letter in s:
    print(letter)
    if letter == 'e' or letter == 's':
        break
        print("out of for loop")
        print()
```

```python
# using while loop
i = 0
while True:
    print(s[i])
    if s[i] == 'e' or s[i] == 's':
        break
        i += 1
        print("out of while loop")
    
```
